**INTRODUCTION**
</center>


<div align="justify"><div style="font-size: 12px;"> Wildfires are among the most destructive natural hazards on Earth, reshaping landscapes in a matter of hours and leaving lasting scars on ecosystems. Beyond the immediate flames, their impact ripples through vegetation, soil, air quality, and human livelihoods, often triggering long-term ecological changes. The Eaton Fire in Los Angeles, which burned thousands of hectares in January 2025, and led to the death of approximately 19 people is a prime example of how rapidly spreading fires can lead to vegetation loss, soil degradation, and long-term landscape alteration.

<div style="font-size: 12px;"> Understanding the severity of these changes is critical for recovery planning and risk mitigation. Traditional ground surveys, while accurate, are time-consuming and limited in spatial coverage. Remote sensing offers a rapid, cost-effective, and scalable solution for monitoring fire impacts across vast areas. Among the available Earth observation platforms, Sentinel-2 stands out for its high-resolution multispectral imagery, enabling detailed analysis of vegetation and soil conditions after fire events (Filipponi, 2019).

<div style="font-size: 12px;"> Remote sensing remains a powerful tool for assessing wildfire changes and in producing data needed to make decisions. One of the most widely used indicators for fire assessment is the Normalized Burn Ratio (NBR), which highlights fire-induced changes in vegetation structure. To capture the true magnitude of these changes, the Differenced Normalized Burn Ratio (dNBR) compares conditions before and after the fire (Key & Benson, 2006). However, raw dNBR values alone do not convey severity levels. To bridge this gap, the U.S. Geological Survey (USGS) developed a classification system that translates dNBR values into meaningful categories, unburned, low, moderate, and high severity, based on field-calibrated Composite Burn Index (CBI) scores.

<div style="font-size: 12px;">This study aims to assess the environmental impact of the Eaton Fire using Sentinel-2 imagery by computing NBR and dNBR, applying USGS burn severity classification thresholds, and estimating the spatial extent of burned areas. The results are presented both numerically and visually, including bar charts showing the distribution of burn severity in hectares and pie charts illustrating the percentage contribution of each severity class. By combining quantitative analysis with clear visual representation, this research provides actionable insights for post-fire recovery planning and demonstrates the value of remote sensing in wildfire assessment.

**DATASET CHARACTERISTICS**
<div style="font-size: 12px;">
The analysis was based on Sentinel-2 Level-2A imagery, accessed through Google Earth Engine (GEE) and processed in Google Colab using Python. Key characteristics of the dataset include:

<div style="font-size: 12px;">

<div style="font-size: 12px;">

<div style="font-size: 12px;">
- Temporal Coverage:

Post-fire period: February 1, 2025  April 30, 2025
<div style="font-size: 12px;">

**PREPROCESSING**
<div style="font-size: 12px;">
All data processing and visualization were performed in Google Colab using Python scripts. Sentinel-2 imagery was imported from GEE, filtered by the defined area of interest (AOI) and cloud percentage (<20%).
The AOI was specified using bounding coordinates [-118.1, 34.15, -118.0, 34.25], which cover the region affected by the Eaton Fire. Two time windows were selected to represent pre-fire and post-fire conditions, ensuring minimal seasonal variation and accurate comparison. To minimize the influence of clouds and cirrus on the analysis, a cloud masking function was applied to Sentinel-2 imagery using the QA60 band. The function identifies cloud and cirrus pixels based on bitwise operations and removes them from the dataset.

**IMAGE IMPORT AND COMPOSITE GENERATION**

<div style="font-size: 12px;">
Sentinel-2 images were imported from GEE into Colab, filtered by AOI and cloud percentage (<20%). The number of images for each period was counted (8 pre-fire and 40 post-fire), and mean composites were generated for analysis. Using mean composites reduces noise and provides a representative image for each time period.

**BURN INDICES COMPUTATION AND BURN SEVERITY CLASSIFICATION**
<div style="font-size: 12px;">
Burn indices were then computed. The Normalized Burn Ratio (NBR) was calculated for both pre-fire and post-fire composite. To quantify the magnitude of change caused by the fire, the Differenced Normalized Burn Ratio (dNBR) was also computed for.
This index highlights areas with significant spectral changes due to burning.
The dNBR values were classified into burn severity categories based on U.S. Geological Survey (USGS) thresholds: unburned (<0.1), low severity (0.10.27), moderate severity (0.270.44), and high severity (>0.44). This classification translates raw dNBR values into meaningful severity levels for interpretation.

**AREA CALCULATION**
<div style="font-size: 12px;">
To quantify the spatial extent of each severity class, pixel areas were summed and converted from square meters to hectares. The total burned area was computed by summing all severity classes except the Unburned / Low Severity category. Percentages for each severity class were calculated relative to the total burned area.

```python
# Authenticate Earth Engine
import ee
ee.Authenticate()
```

```python
#initializing the project
ee.Initialize(project='assignmenttesting-477714')
```

```python
#Defining my Area of Interest
Area_of_Interest = ee.Geometry.Rectangle([-118.1, 34.15, -118.0, 34.25])
```

```python
# Define time periods before and after the Eaton Fire (occurred January 2025)
Pre_fire_start = '2024-12-15'
Pre_fire_end = '2025-01-06'
Post_fire_start = '2025-02-01'
Post_fire_end = '2025-04-30'
```

```python
# Cloud masking function
def mask_s2_clouds(image):
    qa = image.select('QA60')
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11
    mask = (
        qa.bitwiseAnd(cloud_bit_mask).eq(0)
        .And(qa.bitwiseAnd(cirrus_bit_mask).eq(0))
    )
    return image.updateMask(mask).divide(10000)

```

```python
#To load Sentinel 2 dataset for the Pre-Fire Image of my Area of Interest
Pre_fire_S2 = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
          .filterDate(Pre_fire_start, Pre_fire_end)
          .filterBounds(Area_of_Interest)
          .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
          .map(mask_s2_clouds))
```

```python
#To load Sentinel 2 dataset for the Post-Fire Image of my Area of Interest
Post_fire_S2 = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
          .filterDate(Post_fire_start, Post_fire_end)
          .filterBounds(Area_of_Interest)
          .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
          .map(mask_s2_clouds))
```

```python
# Count the number of Sentinel-2 images for the pre-fire period
num_s2_pre_fire = Pre_fire_S2.size().getInfo()


# Count the number of Sentinel-2 images for the post-fire period
num_s2_post_fire = Post_fire_S2.size().getInfo()

```

```python
#CALCULATE NBR FOR PRE & POST FIRE


def compute_nbr(image):
    return image.normalizedDifference(['B8', 'B12']).rename('NBR')

nbr_pre = compute_nbr(Pre_fire_S2.mean())
nbr_post = compute_nbr(Post_fire_S2.mean())

#Calculate the NBR
delta_nbr = nbr_pre.subtract(nbr_post).rename('dNBR')
```

```python
# USGS Burn Severity Classification
classified = (delta_nbr
              .where(delta_nbr.lt(0.1), 0)
              .where(delta_nbr.gte(0.1).And(delta_nbr.lt(0.27)), 1)
              .where(delta_nbr.gte(0.27).And(delta_nbr.lt(0.44)), 2)
              .where(delta_nbr.gte(0.44), 3)
              .rename('BurnSeverity'))

severity_vis = {'min': 0, 'max': 3, 'palette': ['green', 'yellow', 'orange', 'red']}
```

```python
# Compute area per severity class
pixel_area = ee.Image.pixelArea()
area_img = pixel_area.addBands(classified)
stats = area_img.reduceRegion(
    reducer=ee.Reducer.sum().group(groupField=1, groupName='severity'),
    geometry=Area_of_Interest,
    scale=30,
    maxPixels=1e13
)

# Get the raw results
stats_list = stats.get('groups').getInfo()

# Convert 'sum' from square meters to hectares for each item
hectares_stats = []
for item in stats_list:
    item['sum_hectares'] = item['sum'] / 10000
    hectares_stats.append(item)
```

```python
#creating a barchart for the burned severity
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract data from the Earth Engine dictionary
# The result from stats.get('groups') is an Earth Engine List of Dictionaries
# Each dictionary has 'severity' and 'sum' (area) keys.
stats_list = stats.get('groups').getInfo()

# Prepare lists for DataFrame creation
severity_labels = []
areas = []

# Define the labels for each severity class to match the legend
severity_mapping = {
    0: 'Unburned / Low Severity (<0.1)',
    1: 'Low Severity (0.1 - 0.27)',
    2: 'Moderate Severity (0.27 - 0.44)',
    3: 'High Severity (>0.44)'
}

for item in stats_list:
    severity_code = item['severity']
    # Use .get() with a default value to handle cases where a severity code might be missing
    severity_labels.append(severity_mapping.get(severity_code, f'Unknown Severity {severity_code}'))
    areas.append(item['sum'])

# Create a pandas DataFrame
df_severity_area = pd.DataFrame({
    'Severity': severity_labels,
    'Area (m)' : areas
})

# Convert Area from square meters to hectares (1 hectare = 10,000 m)
df_severity_area['Area (hectares)'] = df_severity_area['Area (m)'] / 10000

# Sort the DataFrame by severity order for better visualization
# This assumes the order in severity_mapping is the desired order for the chart
sort_order = [severity_mapping.get(i) for i in sorted(severity_mapping.keys())]
df_severity_area['Severity'] = pd.Categorical(df_severity_area['Severity'], categories=sort_order, ordered=True)
df_severity_area = df_severity_area.sort_values('Severity')

# Define the same colors as used in the map legend (converted to hex)
chart_colors = [
    '#00FF00',  # Green for Unburned / Low Severity
    '#FFFF00',    # yellow
    '#FFA500',    # orange
    '#FF0000'       # red
]
```

```python
#percentage of total burned area
total_burned_area = df_severity_area['Area (hectares)'].sum()
df_severity_area['Percentage'] = (df_severity_area['Area (hectares)'] / total_burned_area) * 100
```

```python
#Creating a pie chart for the burn severity
import matplotlib.pyplot as plt

# Define the custom formatting function for pie chart labels
def autopct_format_hectares(pct):
    total_area_hectares = df_severity_area['Area (hectares)'].sum()
    absolute_area = (pct / 100) * total_area_hectares
    return f'{pct:.1f}% ({absolute_area:.1f} ha)'


```

```python
#Calculating the total burned areas
import pandas as pd

# Filter out the 'Unburned / Low Severity' class
burned_areas_df = df_severity_area[df_severity_area['Severity'] != 'Unburned / Low Severity (<0.1)']

# Calculate the total burned area
total_burned_area_hectares = burned_areas_df['Area (hectares)'].sum()
```
